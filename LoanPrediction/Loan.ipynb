```python
LoanAmountMean = df1['LoanAmount'].mean()
Loan_Amount_Term_Mean = df1['Loan_Amount_Term'].mean()
df1['LoanAmount'] = df1['LoanAmount'].fillna(LoanAmountMean)
df1['Loan_Amount_Term'] = df1['Loan_Amount_Term'].fillna(Loan_Amount_Term_Mean)
df1['Credit_History'] = df1['Credit_History'].fillna(1.0)
```

```python
LoanAmountMean = df2['LoanAmount'].mean()
Loan_Amount_Term_Mean = df2['Loan_Amount_Term'].mean()
df2['LoanAmount'] = df2['LoanAmount'].fillna(LoanAmountMean)
df2['Loan_Amount_Term'] = df2['Loan_Amount_Term'].fillna(Loan_Amount_Term_Mean)
df2['Credit_History'] = df2['Credit_History'].fillna(1.0)
```

```python
df1["Gender"] = df1["Gender"].fillna(df1["Gender"].mode()[0])
df1["Married"] = df1["Married"].fillna(df1["Married"].mode()[0])
df1["Dependents"] = df1["Dependents"].fillna(df1["Dependents"].mode()[0])
df1["Self_Employed"] = df1["Self_Employed"].fillna(df1["Self_Employed"].mode()[0])
```

```python
df2["Gender"] = df2["Gender"].fillna(df2["Gender"].mode()[0])
df2["Married"] = df2["Married"].fillna(df2["Married"].mode()[0])
df2["Dependents"] = df2["Dependents"].fillna(df2["Dependents"].mode()[0])
df2["Self_Employed"] = df2["Self_Employed"].fillna(df2["Self_Employed"].mode()[0])
```

```python
# ApplicantIncome + CoapplicantIncome = TotalIncome
# Make a new column TotalIncome dropping the above two columns.

df1["TotalIncome"] = df1["ApplicantIncome"] + df1["CoapplicantIncome"]
cols = ["ApplicantIncome", "CoapplicantIncome", "Loan_ID"]
df1.drop(cols, axis = 1, inplace = True)
```

```python
df2["TotalIncome"] = df2["ApplicantIncome"] + df2["CoapplicantIncome"]
cols = ["ApplicantIncome", "CoapplicantIncome", "Loan_ID"]
df2.drop(cols, axis = 1, inplace = True)
```
